## Grey's Metric - Fire Incidents

Grey chose to look at which neighborhoods had the lowest amount of reported fire incidents to determine bestness. By his metric, the lower the number of fires the better the neighborhood would rank. You can see in his graph below that the best neighborhoods have only 5-12 fire incidents reported, as opposed to the highest in the dataset being more than 250.

```python
import pandas as pd
fire = pd.read_csv("fire-Incidents.csv")
neighborhood_counts = fire['neighborhood'].value_counts()
bottom_neighborhoods = neighborhood_counts.tail(20) 
bottom_neighborhoods.plot(kind='bar')
```

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
data = pd.read_csv("smart-trash-containers.csv")
top_20 =  data['neighborhood'].value_counts().head(20)
graph = top_20.plot(kind='bar', figsize=(12,10))
graph.set_ylabel('Number of Trash Containers')
graph.set_xlabel('Neighborhoods')
graph.set_title('Smart Trash Containers per Neighborhood')
plt.show()
```

## Combined data - Ratio of Police/Fire Incidents to Number of Smart Trash Containers

The combined metric we chose to use was a combination of the 3, meaning that the best neighborhood would have a low number of crimes reported, a low number of incidents, and a high number of smart trash containers. In order to achieve this we added the fire and police values and divided by the number of trash cans. this provided us with a score which would accurately reflect the true bestness of a neighborhood given our criteria and is pictured in the graph below.

```python
import matplotlib.pyplot as plt
import pandas as pd

# read the files
fire = pd.read_csv("fire-Incidents.csv")
police = pd.read_csv("police.csv")
trash = fire = pd.read_csv("smart-trash-containers.csv")

# count the amount in each neighborhood
fire_counts = fire['neighborhood'].value_counts()
police_counts = police['INCIDENTNEIGHBORHOOD'].value_counts()
trash_counts = trash['neighborhood'].value_counts()
#create metric
composite_metric = (fire_counts + police_counts) / trash_counts

#DataFrame with the results
neighborhood_scores = pd.DataFrame({
    'Neighborhood': composite_metric.index,
    'Composite Metric': composite_metric.values
})

# Sorted neighborhoods based on our metric
top_neighborhoods = neighborhood_scores.sort_values(by='Composite Metric')

# made the graph look nice 
top_neighborhoods.plot(kind='bar', x='Neighborhood', y='Composite Metric', legend=False, figsize=(12,10))
plt.xlabel('Neighborhood')
plt.ylabel('Bestness')
plt.title('Top 20 Neighborhoods Based on bestness Metric')
plt.show()
```
