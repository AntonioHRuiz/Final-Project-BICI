# Smart Containers in the City of Pittsburgh

This is a plot of the amount of smart containers per neighboorhood in pittsburgh. This will then be compared with
the amount of police/fire incidents to determin what the best neighboorhood is

#### First things first, you have to import all of the differnt libraries

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

#### Next, lets get our data from our dataset and put it into a variable to read it

```python
data = pd.read_csv("smart-trash-containers.csv")

data.head(10)
```

#### Now that we have the data lets count the amount of containers there are per neighborhood:

```python
data['neighborhood'].value_counts().head(20)
```

#### Finally, lets plot!

```python
top_20 =  data['neighborhood'].value_counts().head(20)

graph = top_20.plot(kind='bar', figsize=(12,10))
graph.set_ylabel('Number of Trash Containers')
graph.set_xlabel('Neighborhoods')
graph.set_title('Smart Trash Containers per Neighborhood')
plt.show()
```

#### There are so many nighborhoods in pittsburgh that it would be basically impossible to read any graph! This is why we limit the amount to just the top twenty. It could be argued that there is no such thing as the best neighborhood without it having at least 20 smart trash cans.

Therefore, after plotting the amount of smart trash cans per neighborhood it can be deduced that the best neighborhood would be the neighborhood with the most amount of smart trash containers. Therefore, Shadyside would be considered the best neighborhood in the city of Pittsburgh!
